```python
#  required columns: listing_id, host_since, neighbourhood, city, accommodates, price
paris_listings = listings.query("city == 'Paris'").loc[:, ["listing_id", "host_since", "neighbourhood", "city", "accommodates", "price"]]
paris_listings.info()
```

```python
paris_listings = paris_listings[paris_listings["price"] != 0]
paris_listings.info()
```

```python
paris_listings_neighbourhood = paris_listings.groupby("neighbourhood").agg({"price": "mean"}).sort_values(by = "price", ascending = False)
paris_listings_neighbourhood.describe()
```

```python
paris_listings_neighbourhood
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (25, 10))
sns.barplot(data = paris_listings_neighbourhood ,
            x = "price",
            y = paris_listings_neighbourhood.index,
            palette = "Set2", color = "#00a699")

plt.title("PARIS: Average Price/Neighbourhood (Premium on Top)")
plt.xlabel("Price in Euros")
plt.ylabel("AirBnB Neighbourhood Locations.")
plt.grid(True)
plt.show()
```

```python
# ax part is by GPT for tooltips:
plt.figure(figsize = (25, 10))
ax = sns.barplot(data = paris_listings_neighbourhood ,
            x = "price",
            y = paris_listings_neighbourhood.index,
            palette = "Set2", color = "#00a699")


# Add labels to each bar
for i, v in enumerate(paris_listings_neighbourhood["price"]):
    ax.text(v + 5, i, f"€{v:.2f}", color='black', va='center')

plt.title("PARIS: Average Price/Neighbourhood (Premium on Top)")
plt.xlabel("Price in Euros")
plt.ylabel("AirBnB Neighbourhood Locations.")
plt.grid(True)
plt.show()
```

```python
paris_listings["year_of_listing"] = paris_listings["host_since"].dt.year
paris_listings.head()
```

```python
paris_listings.info()
```

```python
paris_listings_over_time = paris_listings.groupby("year_of_listing").agg({
    "listing_id": "count","price": "mean"})
paris_listings_over_time
```

```python
plt.figure(figsize = (25, 7))
sns.lineplot(data = paris_listings_over_time["listing_id"],
             color = "#00a699")

plt.title("PARIS: Listings Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Listings")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (25, 7))
sns.lineplot(data = paris_listings_over_time["price"],
             color = "#00a699")

plt.title("PARIS: Average Price Over Time")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (25, 7))
ax.plot(paris_listings_over_time.index, paris_listings_over_time["listing_id"], color = "#00a699")
ax2 = ax.twinx()
ax2.plot(paris_listings_over_time.index, paris_listings_over_time["price"], color = "#FF5A5F")

ax2.set_ylim(0)

ax.set_xlabel("Year")
ax.set_ylabel("Number of Listings")
ax2.set_ylabel("Average Price")

ax.grid(True)
ax2.grid(True)

plt.show()
```

```python
# Legend from GPT:
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dual-axis plot
fig, ax1 = plt.subplots(figsize=(25, 7))
ax2 = ax1.twinx()

# Plotting on ax1: Listings
line1 = sns.lineplot(
    data=paris_listings_over_time["listing_id"],
    color="#00a699",
    ax=ax1,
    label="Number of Listings"
)

# Plotting on ax2: Price
line2 = sns.lineplot(
    data=paris_listings_over_time["price"],
    color="#FF5A5F",
    ax=ax2,
    label="Average Price"
)

# Set y-limits
ax.set_ylim(bottom=0)
ax2.set_ylim(bottom=0)

# Labels
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of Listings", color="#00a699")
ax2.set_ylabel("Average Price", color="#FF5A5F")

# Grid
ax1.grid(True)
ax2.grid(False)  # Optional: avoid double grids

# Combine both legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

plt.title("AirBnB Listings & Price Trend Over Time")
plt.tight_layout()
plt.show()
```
